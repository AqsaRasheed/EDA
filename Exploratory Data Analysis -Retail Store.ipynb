**Datset** **Loading**

```python
import pandas as pd

# Load the dataset
url = '/content/sample_data/SampleSuperstore.csv'
df = pd.read_csv(url)

# Display the first few rows of the dataset
df.head()
```

**Data Cleaning and Statistics**

```python
# Check for missing values
print(df.isnull().sum())

# Handle duplicates
df = df.drop_duplicates()

# Ensure data types are appropriate
df.dtypes
```

```python
# Generate summary statistics for numerical columns
df.describe()
```

**Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical columns
numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
df[numerical_cols].hist(bins=15, figsize=(15, 6), layout=(2, 2))
plt.show()

# Box plots to check for outliers
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
for i, col in enumerate(numerical_cols):
    sns.boxplot(y=df[col], ax=axes[i])
plt.show()
```

**Profit Analysis**

```python
# Distribution of Profit
sns.histplot(df['Profit'], bins=30, kde=True)
plt.title('Profit Distribution')
plt.show()

# Identify categories with low profit
low_profit = df.groupby('Category')['Profit'].sum().sort_values().head()
low_profit.plot(kind='bar')
plt.title('Categories with Low Profit')
plt.show()
```

**Segment Analysis**

```python
# Compare different segments
segment_profit = df.groupby('Segment')['Profit'].sum()
segment_profit.plot(kind='bar')
plt.title('Profit by Segment')
plt.show()
```

**Category and sub-category Analysis**

```python
# Category analysis
category_profit = df.groupby('Category')['Profit'].sum()
category_profit.plot(kind='bar')
plt.title('Profit by Category')
plt.show()

# Sub-Category analysis
subcategory_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()
subcategory_profit.plot(kind='bar', figsize=(10, 6))
plt.title('Profit by Sub-Category')
plt.show()
```

**Region/State Analysis**

```python
# Region analysis
region_profit = df.groupby('Region')['Profit'].sum()
region_profit.plot(kind='bar')
plt.title('Profit by Region')
plt.show()

# State analysis
state_profit = df.groupby('State')['Profit'].sum().sort_values()
state_profit.plot(kind='bar', figsize=(15, 8))
plt.title('Profit by State')
plt.show()
```

```python
# Analyze how discounts affect sales and profit
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.show()

sns.scatterplot(x='Discount', y='Sales', data=df)
plt.title('Discount vs Sales')
plt.show()
```

Business Problems Identified are

**Low Profitability Categories:** Identify categories and sub-categories with low profitability and investigate reasons (e.g., high discounts, low sales volume).

**Underperforming Segments:** Analyze why certain segments (e.g., Consumer, Corporate, Home Office) may be underperforming.

**Regional Performance:** Regions and states with lower profits may require targeted marketing or operational improvements.

**Impact of Discounts: H**igh discounts might not always lead to higher profits; optimize discount strategies.

**Inventory Management:** Adjust inventory for underperforming products to reduce holding costs and improve cash flow.

By performing this analysis, we can uncover insights and make data-driven decisions to enhance profitability.
